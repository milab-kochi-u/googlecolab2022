## 条件分岐




プログラムは，基本的には上から順に実行されますが，条件分岐（if文）を使うことで，条件によって処理を切り替えることができます。

まずは，条件を満たすかどうかを判定する際に良く用いられる比較演算子を確認してみましょう。

````
#　比較演算子の使用例
a = 3 #@param {type:"integer"}
b = 5 #@param {type:"integer"}
print(a == b) # a　と b が等しければ True そうでなければ False
print(a != b) # a　と b が異なれば True そうでなければ False
print(a < b) # a より b が大きければ True そうでなければ False
print(a <= b) # a が b 以下であれば True そうでなければ False
print(a > b)  # a が b より大きければ True そうでなければ False
print(a >= b) # a が b 以上であれば True そうでなければ False
````

条件分岐を行う if 文は，
```
if 条件式:
  条件式が True のときに実行するプログラム
```
のように記述します。

「条件式が True のときに実行するプログラム」は，インデント（字下げ）されていることに注意してください。インデントされている行が「条件式が True のときに実行する範囲」とみなされます。

else を使って，「条件式を満たさない（False）ときに実行するプログラム」も書くことができます。
```
if 条件式:
  条件式が True のときに実行するプログラム
else:
  条件式が False のときに実行するプログラム
```

````
# 成績判定
score = int(input('得点 = '))
if score >= 60:
  print('合格')
else:
  print('不合格')
````

````
# 偶数・奇数の判定
n = 3 #@param {type:"integer"}
if n % 2 == 0:  # n を 2　で割った余りが 0 なら偶数，1なら奇数
  print('偶数')
else:
  print('奇数')
````

インデントを重ねることで，入れ子構造の if 文を書くこともできます。これにより，複雑な条件分岐も可能です。

````
# 成績判定 ver.2
score = int(input('得点 = '))
if score >= 60:
  if score >= 70:
    if score >= 80:
      if score >= 90:
        print('秀')
      else:
        print('優')
    else:
      print('良')
  else:
    print('可')
else:
  print('不可')
````

上の ver.2 のプログラムは，少し読みづらいので，下の ver.2.5 のように書き換えてみました。同じように動作します。

````
# 成績判定 ver.2.5
score = int(input('得点 = '))
if score >= 90:
  print('秀')
else:
  if score >= 80:
    print('優')
  else:
    if score >= 70:
      print('良')
    else:
      if score >= 60:
        print('可')
      else:
        print('不可')
````

上のように，else に続けて if 文を書く場合は，elif で置き換えられるかもしれません。

elif を使えば，下の ver.3 のように，インデントが深くならずにすっきりとしたプログラムになります。

もちろん，これも ver.2 や ver.2.5 と同様に動作します。

````
# 成績判定 ver.3
score = int(input('得点 = '))
if score >= 90:
  print('秀')
elif score >= 80:
  print('優')
elif score >= 70:
  print('良')
elif score >= 60:
  print('可')
else:
  print('不可')
````

## リスト（配列）




繰り返し処理のプログラムの書き方を学ぶ前に，リスト（配列）について紹介しておきます。

先週，データをいれる容れ物として「変数」を紹介しました。

1つの変数には，基本的には1つのデータ（値）が入るのですが，リスト（配列）を使うと，複数のデータ（値）を持つことができます。

次のプログラムでは，4つのデータが入った prefs と populations というリスト型の変数が使われています。

「1番目のデータ」が欲しい場合は，`prefs[0]`のように指定します。何番目かを示す数字は，**0から始まる**ことに注意しましょう。

````
prefs = ['徳島', '香川', '愛媛', '高知']
populations = [743000, 967000, 1364000, 714000]
print(prefs)
print(prefs[0], populations[0])
print(prefs[3], populations[3])
n = 2 #@param {type:"integer"}
print(prefs[n], populations[n])
````

辞書型や連想配列などと呼ばれる次のような書き方を用いると，「何番目のデータ」というように数字で指定するのではなく，名前（ラベル）で指定することも可能です。

````
pref_populations = { '徳島': 743000, '香川': 967000, '愛媛': 1364000, '高知': 714000 }
print(pref_populations['高知'])
````

これらを組み合わせて，複雑なデータ構造を表現することもできます。

````
prefectures = [
  { 'name': '徳島', 'capital': '徳島', 'population': 743000 },
  { 'name': '香川', 'capital': '高松', 'population': 967000 },
  { 'name': '愛媛', 'capital': '松山', 'population': 1364000 },
  { 'name': '高知', 'capital': '高知', 'population': 714000 }
]
n = 2 #@param {type:"integer"}
print(prefectures[n])
print(prefectures[n]['capital'])
````

また，反復した数字のリストであれば，range関数とlist関数を組み合わせて作ることができます。

````
print(list(range(10)))
print(list(range(5, 11)))
a = list(range(1, 20, 3))
print(a)
print(a[2])
````
